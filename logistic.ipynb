```python
import pandas as pd
import chart_studio.plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from scipy import optimize
from sklearn.linear_model import LogisticRegression

def Sigmoid(x): 
    return 1.0/(1.0 + np.exp(-x))
```

```python
iris_path = 'iris.csv'
data = pd.read_csv(iris_path)
x1 = np.array(data.iloc[:,0])
x2 = np.array(data.iloc[:,1])
x3 = np.array(data.iloc[:,2])
x4 = np.array(data.iloc[:,3])
```

```python
X = data.iloc[0:99, 0:4]
xMat = np.mat(X)
Y = data.iloc[0:99, 4]
yMat = np.mat(Y)
x1 = X.iloc[:, 0]
x2 = X.iloc[:, 1]
x3 = X.iloc[:, 2]
x4 = X.iloc[:, 3]

a = 1 * xMat[:, 0]
b = 2 * xMat[:, 1]
c = 3 * xMat[:, 2]
d = 4 * xMat[:, 3]
# print(a + b +c + d)
```

```python
def J (theta, x = xMat, y = yMat):
    a = theta[0] * xMat[:, 0]
    b = theta[1] * xMat[:, 1]
    c = theta[2] * xMat[:, 2]
    d = theta[3] * xMat[:, 3]
    temp = a + b + + c + d
    
    # print (temp)
    p = Sigmoid(temp)
    return -np.sum(y * np.log(p) + (1-y) * np.log(1-p))/100
```

```python
theta = optimize.fmin_bfgs(J, [0,0,0,0], disp=False)
print (theta)
```

```python
temp = theta[0] * X.iloc[:,0] + theta[1] * X.iloc[:, 1] + theta[2] * X.iloc[:, 2] + theta[3] * X.iloc[:, 3]
pred = Sigmoid(temp)
plt.scatter([temp[0:49]], [pred[0:49]], c='red', alpha=0.2)
plt.scatter([temp[50:99]], [pred[50:99]], c='green', alpha=0.2)
# plt.scatter([temp[100:149]], [pred[100:149]], c='blue', alpha=0.2)
xfit = np.linspace(-50, 50, 200)
plt.plot(xfit, Sigmoid(xfit), c='black', alpha=.7)
```
